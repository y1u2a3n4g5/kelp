```python
import matplotlib.pyplot as plt
import tifffile

# Load the image
img_path = "C:\\Users\\guoyy\\Downloads\\test_features.tar_cVzb1K0\\test_satellite\\EC437488_satellite.tif"
with tifffile.TiffFile(img_path) as tif:
    img = tif.asarray()

# Assuming img.shape is (350, 350, 7)
n_channels = img.shape[2]

# Create a figure to hold the subplots
plt.figure(figsize=(15, 8))  # Adjust the size as needed

# Loop through each channel
for i in range(n_channels):
    plt.subplot(1, n_channels, i + 1)  # Create a subplot for each channel
    plt.imshow(img[:, :, i])  # Display as grayscale
    plt.title(f'Channel {i+1}')
    plt.axis('off')  # Hide the axis for clarity

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import tifffile

# Load the first image (single-channel or RGB)
img_path_kelp = "C:\\Users\\guoyy\\Downloads\\train_labels.tar_l8u2RP0\\train_kelp\\DN412676_kelp.tif"
with tifffile.TiffFile(img_path_kelp) as tif:
    img_kelp = tif.asarray()

# Load the second image (multi-channel)
img_path_satellite = "C:\\Users\\guoyy\\Downloads\\train_features.tar_MLIC14m\\train_satellite\\DN412676_satellite.tif"
with tifffile.TiffFile(img_path_satellite) as tif:
    img_satellite = tif.asarray()
n_channels = img_satellite.shape[2]

# Assuming img_satellite.shape is (350, 350, 7)
# Setup the figure and axes
plt.figure(figsize=(20, 10))  # Adjust the size as needed

# Add the single image as the first subplot
plt.subplot(2, 4, 1)  # Consider it's placed in a 2x4 grid
plt.imshow(img_kelp)  # Assuming it's grayscale; adjust as needed
plt.title('Kelp Image')
plt.axis('off')

# Loop through each channel of the satellite image and add as subsequent subplots
for i in range(n_channels):
    plt.subplot(2, 4, i + 2)  # Offset by 2 to account for the first kelp image
    plt.imshow(img_satellite[:, :, i], cmap='gray')  # Display each channel in grayscale
    plt.title(f'Channel {i+1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
